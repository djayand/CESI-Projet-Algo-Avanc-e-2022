## Definition and constraints

First, we have been asked to compute a tour with a single vehicle, which corresponds to the Travelling salesman problem (TSP).

In our case, we will have to find routes to make delivery rounds to reach different cities on a large territory, we then admit that the graph is complete, and not oriented. The objective is to pass only once through all the vertices, and to return to the starting point, minimizing the total distance traveled. That is to say, find a halmitone cycle for the given graph. 


- Multiple tour

We have chosen to add the management of several vehicles, using k trucks, thus transforming the TSP into a Vehicle Routing Problem (VRP). The VRP is a generalization of the TSP because each instance of the TSP can be considered as a particular instance of the VRP with a single vehicle.


- Volume management

We have chosen to limit the volume of the trucks, which means adding the capacity constraint (also called CVRP for Capacitated Vehicle Routing Problem): The vehicles have a limited carrying capacity (quantity, size, weight, etc.). 
The CVRP is modeled as a complete undirected graph G = (V, A), where V = {v0, v1, ..., vn} is the set of vertices such that v0 denotes the depot and v1, ..., vn represent the customers. The objective of the CVRP problem is therefore to minimize the total distance traveled such that: each tour must start and end at the depot, the total load of each tour must not exceed the vehicle capacity, each customer must be served by one and only one vehicle. All nodes are linked together, which means that each city can be visited from all other cities.

The objectives are:

- Minimization of the total distance covered by the vehicles
- Minimization of the total duration of the circuits



## Complexity


The VRP is a very difficult problem to solve in an optimal way, this difficulty and its numerous applications make it one of the most interesting and studied. A combinatorial problem is a problem whose goal is to find the optimal combination in a set of possible combinations.

The VRP, as well as the TSP, belong to the so-called NP-complete problems. NP problems are problems for which we can efficiently check if a solution is valid and for which the possible solutions can be efficiently generated in a non-deterministic way. NP-complete problems are the most difficult problems in the NP class, indeed, if an efficient algorithm were known for any NP-complete problem, then all other NP problems could be solved efficiently.

The problems could be solved efficiently. The known algorithms for solving VRPs have exponential worst-case time complexity. This means that there are instances for which the algorithm will take time exponential to the size of the instance.
